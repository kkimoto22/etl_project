```python
# Import dependencies
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
from sqlalchemy import func
```

```python
# Store first CSV into a dataframe
business_licenses = "business_licenses/business-licenses.csv"
business_licenses_df = pd.read_csv(business_licenses, error_bad_lines=False, index_col=False, dtype='unicode')
business_licenses_df.head()
```

```python
# Store second CSV into a dataframe
sidewalk_cafe = "sidewalk_cafe/sidewalk-cafe-permits.csv"
sidewalk_cafe_df = pd.read_csv(sidewalk_cafe, error_bad_lines=False, index_col=False, dtype='unicode')
sidewalk_cafe_df.head()
```

```python
# Remove 
sidewalk_unique = sidewalk_cafe_df.drop_duplicates('LEGAL NAME')
sidewalk_unique.head()
```

```python
sidewalk_unique.count()
```

```python
rev_sidewalk = sidewalk_unique[['ACCOUNT NUMBER', 'LEGAL NAME', 'PAYMENT DATE', 'DOING BUSINESS AS NAME']].copy()
rev_sidewalk.head()
```

```python
business_licenses_unique = business_licenses_df.drop_duplicates('LEGAL NAME')
business_licenses_unique.head()
```

```python
business_licenses_unique.count()
```

```python
rev_business = business_licenses_unique[['ACCOUNT NUMBER', 'LEGAL NAME', 'LICENSE STATUS', 'DOING BUSINESS AS NAME']].copy()
rev_business.head()
```

```python
rev_business.rename({
    'ACCOUNT NUMBER':'account_number', 
    'LEGAL NAME': 'legal_name', 
    'LICENSE STATUS': 'license_status', 
    'DOING BUSINESS AS NAME': 'doing_business_as_name'
}, axis='columns', inplace=True)
```

```python
rds_connection_string = "postgres:<password>@localhost:5432/etl_project"
engine = create_engine(f'postgresql://{rds_connection_string}')
engine.table_names()
```

```python
rev_sidewalk.rename({
    'ACCOUNT NUMBER':'account_number', 
    'LEGAL NAME': 'legal_name', 
    'PAYMENT DATE': 'payment_date', 
    'DOING BUSINESS AS NAME': 'doing_business_as_name'
}, axis='columns', inplace=True)
```

```python
rev_business.to_sql(name='business', con=engine, if_exists='append', index=False)
```

```python
rev_sidewalk.to_sql(name='sidewalk', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('''SELECT * from business
INNER JOIN sidewalk
ON business.account_number = sidewalk.account_number;''', con=engine).head()
```
